```python
#For testing images
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import time

net = cv2.dnn.readNet('/content/drive/MyDrive/yolov3/darknet/cfg/yolov3-tiny.cfg','/content/drive/MyDrive/yolov3/yolov3_training_last.weightss')

classes = []
with open("/content/drive/MyDrive/classes.txt", "r") as f:
    classes = f.read().splitlines()

font = cv2.FONT_HERSHEY_PLAIN
colors = np.random.uniform(0, 255, size=(100, 3))

while True:

    img = cv2.imread("/content/Gao_ship_vh_020161130030501002.jpg")
    height, width,_ = img.shape

    start_time = time.time()
    blob = cv2.dnn.blobFromImage(img, 1/255, (256, 256), (0,0,0), swapRB=True, crop=False)
    net.setInput(blob)
    output_layers_names = net.getUnconnectedOutLayersNames()
    layerOutputs = net.forward(output_layers_names)
    
    end_time = time.time()
    elapsed_time = end_time - start_time
    print('Done! Took {} seconds'.format(elapsed_time))

    boxes = []
    confidences = []
    class_ids = []

    for output in layerOutputs:
        for detection in output:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.2:
                center_x = int(detection[0]*width)
                center_y = int(detection[1]*height)
                w = int(detection[2]*width)
                h = int(detection[3]*height)

                x = int(center_x - w/2)
                y = int(center_y - h/2)

                boxes.append([x, y, w, h])
                confidences.append((float(confidence)))
                class_ids.append(class_id)


    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.2, 0.4)
    if len(indexes)>0:
        for i in indexes.flatten():
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            confidence = str(round(confidences[i],2))
            color = colors[i]
            cv2.rectangle(img, (x,y), (x+w, y+h), color, 2)
            cv2.putText(img, label + " " + confidence, (x, y+20), font, 1, (255,255,255), 2)

    cv2_imshow(img)
    break


cv2.destroyAllWindows()
```
